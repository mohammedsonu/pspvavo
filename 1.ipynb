MEMBERSHIP FUNCTIONS

```python
fruits=["banana","apple","mango","orange"]
fruits.append("grapes")
print(fruits)
food=["idly","dosa","puri"]
print(food)
fruits.extend(food)
fruits.insert(0,"green apple")
print(fruits)
fruits.remove("puri")
print(fruits.pop(0))
print(fruits)
```
