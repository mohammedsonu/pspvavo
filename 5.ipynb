```python
n=int(input("enter the number"))
count=0
for i in range(2,n):
  if (n%i==0):
    count=count+1
if(count==0):
    print("given number is prime numnber")
else:
      print("given number is not a prime number")
```

```python
n=int(input("enter the number"))
evensum=0
oddsum=0
for i in range(1,n+1):
  if(i%2==0):
    evensum=evensum+i
  else:
      oddsum=oddsum+i
print("even sum is:",evensum)
print("odd sum is:",oddsum)
```

```python
n=int(input("enter the number"))
fact=1
for i in range(1,n+1):
  fact=fact*i
print("factorial of %d is %d"%(n,fact))
```
